### Pandas

Pandas is a Python library which makes working with large datasets convenient; it utilizes an object called a DataFrame, which stores data in a CSV like format. A Pandas DataFrame allows you to perform Exploratory Data Analysis and run various models for both supervised and unsupervised learning.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

##### NaN Values in Pandas DataFrames

You cannot run models on a Pandas DataFrame with NaN values. Therefore, NaN values must be dealt with beforehand. 

There are a few ways to deal with NaN values:

1. Imputation of the mean: filling the missing values with the mean of other values within that column. 
    This should only be done if the data has been combed throug for outliers, if there is sufficient data with which to fill the NaN values.
    Keep in mind that imputing data is creating data, and as such, may be distorting a future model.

2. Dropping NaN values: it sucks to have to drop data, but sometimes is necessary (for example, if the column is 90% NaN values.
    
3. Using more advanced methods, like K-Nearest-Neighbors, to fill in the values.
    Not getting in KNN for this class.

```python
# Here's a function I used every single time I have a Pandas df
def get_percentage_missing(df):
    df_with_any_null_values = df[df.columns[df.isnull().any()].tolist()]
    num = df_with_any_null_values.isnull().sum()
    den = len(df_with_any_null_values)

    percent_missing = round(num/den, 2)
    
    return percent_missing
```

```python
get_percentage_missing(wine2)
```

```python
def sns_heatmap(df):#, labels):

        plt.style.use('seaborn-darkgrid')

        # Convert categorical df types into ints so corr will work
        dummy_df = pd.get_dummies(df)
        # Compute the correlation matrix
        corr = dummy_df.corr()

        sns.set(style="white")

        # Generate a mask for the upper triangle
        mask = np.zeros_like(corr, dtype=np.bool)
        mask[np.triu_indices_from(mask)] = True

        # Set up the matplotlib figure
        f, ax = plt.subplots(figsize=(16,16))
        # Generate a custom diverging colormap
        cmap = sns.diverging_palette(220, 10, as_cmap=True)
        # Draw the heatmap with the mask and correct aspect ratio

        # Generage plot. Optional argument mask
        cm = sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
        square=True, linewidths=.5, mask=mask, cbar_kws={"shrink": .5}, annot=True, fmt='.2f', annot_kws={"size":9})#,
        #xticklabels=labels, yticklabels=labels)

        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=45)

        plt.subplots_adjust(left=0.12, bottom=0.21, right=0.86, top=0.88,
                    wspace=0.20, hspace=0.20)
        plt.show()
```

```python
def scatter_plot(df, column1, column2, title):

        plt.style.use('seaborn-darkgrid')

        fig, ax = plt.subplots(figsize=(14, 14))
        ax.scatter(df[column1], df[column2], color=["grey", 'blue'],)
        ax.set_xlabel(column1); ax.set_ylabel(column2)
        plt.style.use('seaborn-darkgrid')
        plt.title(title)

        plt.show()
```

```python
scatter_plot(wine2, 'citric_acid', 'fixed_acidity', 'Citric Acid & Fixed Acidity')
```

```python
 def nested_boxplot(df, column1, column2, title):

        plt.style.use('seaborn-darkgrid')

        sns.set(style="ticks")
        plt.figure(figsize=(15,15))
        sns.boxplot(x=column1, y=column2, data=df, palette='PuBuGn_d')
        sns.despine(offset=10, trim=True)
        plt.xticks(rotation=90)
        plt.title(title)

        plt.show()
```

```python
nested_boxplot(wine2, 'citric_acid', 'fixed_acidity', 'Citric Acid & Fixed Acidity')
```

```python
def histogram(df, column, title):

        plt.style.use('seaborn-darkgrid')

        fig = plt.figure(figsize=(14,14))
        ax = fig.add_subplot(111)
        plt.style.use('seaborn-darkgrid')
        ax.hist(df[column], bins=25, alpha=0.5, normed=True)
        plt.title(title)
        plt.xlabel(column)
        plt.ylabel('counts')

        plt.show()
```

```python
histogram(wine2, 'citric_acid', 'Citric Acid');
```
