```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import poisson
%matplotlib inline
```

# Normal Distribution

```python
distribution = norm(loc=0, scale=1)
xs = np.linspace(-3.0, 3.0, 100)
ys = distribution.pdf(xs)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,4))
ax.plot(xs, ys)
ax.set_ylim(0.0, 0.5)
ax.set_xlabel('x in standard deviations')
ax.set_ylabel('pdf(x)')
ax.set_title('PDF norm(mu=0, sigma=1)')
```

```python
distribution = norm(loc=0, scale=1)
xs = np.linspace(-3.0, 3.0, 100)
ys = distribution.cdf(xs)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,4))
ax.plot(xs, ys)
ax.set_xlabel('x in standard deviations')
ax.set_ylabel('cdf(x)')
ax.set_title('CDF norm(mu=0, sigma=1)')
```

```python
distribution = norm(loc=0, scale=1)
xs = np.linspace(-3.0, 3.0, 100)
pdf_ys = distribution.pdf(xs)
cdf_ys = distribution.cdf(xs)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,4))
ax.plot(xs, pdf_ys, label='PDF', color='b', linestyle=':', linewidth=5)
ax.plot(xs, cdf_ys, label='CDF', color='r', linestyle='-.', linewidth=5)
ax.set_title('CDF and PDF norm(mu=0, sigma=1)')
ax.set_xlabel('x in std. dev.')
ax.legend(loc='best')
```
