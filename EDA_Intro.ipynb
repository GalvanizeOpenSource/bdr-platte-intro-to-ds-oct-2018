```python
# Generate some data (later we'll use real data for plotting)
xs = random.normal(0, 0.1, 50)
ys = random.normal(0, 0.1, 50)

# Establish a canvas
fig = plt.figure(figsize=(10, 10))

# establish a plot on that canvas change (1,1,1) for diff size
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x=xs, y=ys)

# Set the labels of the plot
ax.set_xlabel('xs', size=15)
ax.set_ylabel('ys', size=15);
```

##### Grids of Plots

```python
# Set up some polynomial orders (amount of curve) for demo
orders = range(1, 4)
# Put colour arguments into list
colors = ['r', 'g', 'b']
# Put title arguments into list
titles = ['linear', 'quadratic', 'cubic']

# Establish the x data and thus the X axis range as well
xs = np.linspace(-2., 2., 100)

# Tuple unpacking (explain this)
fig, axes = plt.subplots(nrows=1, ncols=len(orders), figsize=(10,3))

# Just make it nicer to look at
plt.tight_layout()

# Plot all three at once using zip function
for ax, order, color, title in zip(axes, orders, colors, titles):
    # Generate y data
    ys = xs ** order
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('x^{}'.format(order))
    ax.plot(xs, ys, color)
```

##### Multiple Plots on Same Figure

```python
orders = range(1, 4)  # Orders of polynomails
colors = ['r', 'g', 'b']
labels = ['linear', 'quadratic', 'cubic']
markers = ['+', 'o', '*']
lss = ['--', '-', ':']
lws = [6, 3, 2]
alpha = 0.5

xs = np.linspace(-2., 2., 10)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('x')
ax.set_xlabel('y')

for order, color, label, marker, ls, lw in zip(orders, colors, labels, markers, lss, lws):
    ys = xs ** order
    ax.plot(xs, ys, label=label, color=color, marker=marker, markersize=lw * 2, ls=ls, lw=lw, alpha=alpha)

ax.legend(loc='best')
```

##### Random Walk with Matplotlib

```python
number_of_walks = 5
steps_in_each_walk = 100

# Make random distances for each step
step_distances = random.randn(number_of_walks * (steps_in_each_walk -1)).reshape(5, -1)

# Make each walk start at step 0 by horizontally stacking 0s
# at the beginning of the walks

starts = np.zeros(step_distances.shape[0]).reshape(-1, 1)
walk_distances = np.hstack([starts, step_distances])
walks = walk_distances.cumsum(axis=1)

# Now make an array of integers that will serve as the x axis
steps = np.arange(walks.shape[1])

# Find the mean distance at the end of each walk by taking the mean
# of the last column of walks

mean_dist = walks[:, -1].mean()

# Colors, labels, and alpha settings for our lines
colors = ['r', 'g', 'b', 'm', 'c']
labels = ['A', 'B', 'C', 'D', 'E']
alpha = 0.5

# Create a grid of 1 by 1 plots and size it
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(1, 1, 1)

# Make the axis labels
ax.set_ylabel('distance')
ax.set_xlabel('steps')

# Force the x axis limits to match the first and last step
# by default matplotlib may try to pad this which will make
# it look like a mess

ax.set_xlim(steps[0], steps[-1])

# Plot each walk. Zip together the iterables we will need and
# just plot to the same axes

for walk_idx, color, label in zip(range(walks.shape[0]), colors, labels):
    ax.plot(steps, walks[walk_idx], label=label, color=color, lw=1)

# Behind every great plot is an inspirational legend.
# Make your here

ax.legend(loc='best')

# Make a line (and label for it) that show the mean distance
# convered by each walk

ax.axhline(mean_dist, 0, steps[-1])
ax.text(steps[-1] + 1, mean_dist, 'μ={:.2f}'.format(mean_dist));
```

### Plotting Distributions Intro

```python
# Imports we'll need
from scipy.stats import norm
from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import poisson
```

We'll get more into this in Statistics, but here are some high level definitions that will be helpful for the following section:

1. Probability Distribution: A mathematical function that provides the probabilities of occurrence of different possible outcomes in an experiment. In more technical terms, the probability distribution is a description of a random phenomenon in terms of the probabilities of events. For instance, if the random variable X is used to denote the outcome of a coin toss ("the experiment"), then the probability distribution of X would take the value 0.5 for X = heads, and 0.5 for X = tails (assuming the coin is fair). (Thanks, Wiki!)

2. PDF (Probability Density Function):
    For continuous variables only.
    Used to specify the probability of the random variable falling within a particular range of values, as opposed to taking on any one value.

3. PMF (Probability Mass Function):
    For discrete variables only.
    The probability that a discrete random variable is exactly equal to some value.

4. CDF (Cumulative Distribution Function):
    For both continuous and discrete variables.
    The probability that X will take on a value less than or equal to x.
    (Sean uses CMF when refering to CDF for discrete values, so you may see both CDF and CMF, but they are the same.)

5. PPF (Percent Point Function):
    The inverse of the CDF.

##### Normal Distribution

The Normal Distribution describes many natural phenomenon. It's a continuous distribution; it describes the probability that X lies within a particular range under the normal bell curve, which has a mean: mu and a standard deviation: sigma.

```python
# establish the distribution
distribution = norm(loc=0, scale=1)

# Generate data; notice the ys is the pdf of the normal distribution
xs = np.linspace(-3.0, 3.0, 100)
ys = distribution.pdf(xs)

# Establish the canvas and the plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.plot(xs, ys)

# Set the title and labels
ax.set_title('PDF norm(mu=0, sigma=1)', size=20)
ax.set_xlabel('x in standard deviations', size=15)
ax.set_ylabel('pdf(x)', size=15);
```

```python
# Define the distribution
distribution = norm(loc=0, scale=1)

# Generate the data, again notice ys
xs = np.linspace(-3.0, 3.0, 100)
ys = distribution.cdf(xs)

# Establish the canvas and plots
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.plot(xs, ys)

# Title and labels
ax.set_title('CDF norm(mu=0, sigma=1)', size=20)
ax.set_xlabel('x in standard deviations', size=15)
ax.set_ylabel('cdf(x)', size=15);
```

```python
# Define the distribution
distribution = norm(loc=0, scale=1)

# Set x axis
xs = np.linspace(-3.0, 3.0, 100)

# Define both the CDF and the PDF
cdf_ys = distribution.cdf(xs)
pdf_ys = distribution.pdf(xs)

# Establish the canvas and the plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))

# On the plot (ax), plot the pdf and cdf
ax.plot(xs, pdf_ys, color='b', linestyle=':', linewidth=5, label='PDF')
ax.plot(xs, cdf_ys, color='r', linestyle='-.', linewidth=5, label='CDF')

# Set the title and label and legend
ax.set_title('CDF and PDF norm(mu=0, sigma=1)', size=20)
ax.set_xlabel('x in standard deviations', size=15)
ax.legend(loc='best');
```

##### Bernoulli Distribution

A discrete distribution that describes the situation in which X takes on the value 1 with probability, p or the value 0 with probability, q = 1 - p in a single trial or experiment.

```python
# This is my sample space (only 1 and 0 possible outcomes)
xs = [0,1]

# Probability of an outcome
p = 0.1

# Y axis is bernoulli distribution PMF (recall for discrete values)
# Plotting the PMF for each value in xs at a probability of p
pmf_ys = [bernoulli.pmf(k, p) for k in xs]

# Establish canvas and plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))

# Set title, labels, x and y spread 
ax.set_title('Bernoulli PMF, p=0.1')
ax.set_xlabel('outocme (k)')
ax.set_ylabel('probability of outcome')
ax.set_ylim(0.0, 1.0)
ax.set_xlim(-1, 2)

# Add the blue star at the top
ax.scatter(xs, pmf_ys, color='b', alpha=0.5, marker='*', s=250)

# Add the lines to indicate the outcomes
ax.vlines(xs, 0, pmf_ys, color='r', lw=5, alpha=0.5);
```

##### Binomial Distribution

A discrete probability distribution of the number of successes in a sequence of n independent experiments, each asking a yes–no question, and each with its own boolean-valued outcome: a random variable containing a single bit of information: success/yes/true/one (with probability p) or failure/no/false/zero (with probability q = 1 − p). (Notice that this is an extension of the Bernoulli distribution. (So, it basically shows the distribution of successful trials.)

```python
# Establish the number of trials and the probability of success
n = 100
p = 0.1

# Set up the x axis
xs = np.arange(binom.ppf(q=0.01, n=n, p=p), binom.ppf(q=0.99, n=n, p=p))

# Set the ys
pmf_ys = binom.pmf(xs, n=n, p=p)

# Establish the canvas and the plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))

# Add the blue arrows
ax.scatter(xs, pmf_ys, color='b', alpha=1.0, marker='^', s=250)

# Add the green lines
ax.vlines(xs, 0, pmf_ys, color='g', lw=5, alpha=0.5)

# Set title, labels
ax.set_title('Binomial PMF, n=100, p=0.1', size=20)
ax.set_xlabel('Number of successful trials (k)', size=15)
ax.set_ylabel('PMF(k)', size=15);
```

```python
# Number of trials, probability of success
n = 100
p = 0.1

# Set up x axis
xs = np.arange(binom.ppf(q=0.01, n=n, p=p), binom.ppf(q=0.99, n=n, p=p))

# Set up y axis
cmf_ys = binom.cdf(xs, n=n, p=p)

# Establish canvas and plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))

# Add blue triangles
ax.scatter(xs, cmf_ys, color='b', alpha=1.0, marker='^', s=250)

# Add lines
ax.vlines(xs, 0, cmf_ys, color='g', lw=5, alpha=0.5)

# Set title, labels
ax.set_title('Binomial CDF (CMF), n={}, p={}'.format(n, p), size=20)
ax.set_xlabel('Number of successful trials (success <= k)', size=15)
ax.set_ylabel('CMF(k)', size=15);
```

##### Poisson Distribution

A discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant rate and independently of the time since the last event.

```python
# Establish lambda (also mu for Poisson); the rate
mu = 10

# Set x axis
xs = np.arange(poisson.ppf(q=0.001, mu=mu), poisson.ppf(q=0.999, mu=mu), 1)

# Set y axis with PMF
pmf_ys = poisson.pmf(xs, mu=mu)

# Establish canvas, plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))

# Add red dots
ax.scatter(xs, pmf_ys, color='r', alpha=1.0)

# Add lines
ax.vlines(xs, 0, pmf_ys, color='r', lw=5, alpha=0.5)

# Title, labels, y axis range
ax.set_title('Posisson PMF, mu={}'.format(mu), size=20)
ax.set_ylabel('pmf(k)', size=15)
ax.set_xlabel('k', size=15)
ax.set_ylim(0, pmf_ys.max() * 1.1);
```

```python
mu = 10
xs = np.arange(poisson.ppf(q=0.001, mu=mu), poisson.ppf(q=0.999, mu=mu), 1)
cmf_ys = poisson.cdf(xs, mu=mu)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
ax.scatter(xs, cmf_ys, color='r', alpha=1.0)
ax.vlines(xs, 0, cmf_ys, color='r', lw=5, alpha=0.5)
ax.set_title('Posisson CDF (CMF), mu={}'.format(mu), size=20)
ax.set_ylabel('cmf(k)', size=15)
ax.set_xlabel('k', size=15)
ax.set_ylim(0, cmf_ys.max() * 1.1);
```
